```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
from selenium.common.exceptions import NoSuchElementException, ElementNotInteractableException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException

def check_for_warning(driver):
    try:
        elements = driver.find_elements(By.XPATH, '//*[contains(text(), "申込制限回数を超えています")]')
        if elements:
            return True
    except NoSuchElementException:
        return False

def click_nextbtn(driver):
    try:
        nextbtn = driver.find_element(By.CSS_SELECTOR, 'a[onclick="javascript:isSubmitDataSet(\'Enter\');return false;"]')
        nextbtn.click()
        time.sleep(2)
    except (NoSuchElementException, ElementNotInteractableException) as e:
        print("Failed to click the Next button.")
        return False
    return True

def pop_OK(driver):
    try:
        popup_ok = driver.find_element(By.CSS_SELECTOR, f'input[id ="popup_ok"]')
        popup_ok.click()
        time.sleep(2)
    except (NoSuchElementException, ElementNotInteractableException) as e:
        print("Failed to click the Next button.")
        return False
    return True

def kakute(driver):
    try:
        kakutebtn = driver.find_element(By.CSS_SELECTOR, 'a[onclick="javascript:isSubmitDataSetConfirm(\'Enter\');return false;"]')
        kakutebtn.click()
        time.sleep(2)
    except (NoSuchElementException, ElementNotInteractableException) as e:
        print("Failed to click the Next button.")
        return False
    return True

def cyucontinue(driver):
    try:
        cyucontinue = driver.find_element(By.CSS_SELECTOR, 'a[onclick="javascript:isSubmitDataSet(\'ContinueEntry\');return false;"]')
        cyucontinue.click()
        time.sleep(2)
    except (NoSuchElementException, ElementNotInteractableException) as e:
        print("Failed to click the Next button.")
        return False
    return True

def next7(driver):
    try:
        next7btn = driver.find_element(By.CSS_SELECTOR, 'a[onclick="javascript:isSubmitDataSet(\'OffsetNext\');return false;"]')
        next7btn.click()
        time.sleep(2)
    except (NoSuchElementException, ElementNotInteractableException) as e:
        print("Failed to click the Next button.")
        return False
    return True


def doni1521(driver):
    try:
        # 時間帯のテキストリスト
        time_slots = ["15:00 ～ 18:00", "18:00 ～ 21:00"]
        
        for time_slot_text in time_slots:
            try:
                # 時間帯を特定
                time_slot = driver.find_element(By.XPATH, f'//td[contains(text(), "{time_slot_text}")]/..')
                
                # <td> elements with class 'sat' within the specified time slot row
                sat_elements = time_slot.find_elements(By.CLASS_NAME, 'sat')
                
                for sat_element in sat_elements:
                    try:
                        # Find all <input> elements within the <td> element with an id that starts with 'i_record'
                        input_elements = sat_element.find_elements(By.CSS_SELECTOR, 'input[id^="i_record"]')
                        
                        for slot_button in input_elements:
                            try:
                                slot_button.click()
                                slot_id = slot_button.get_attribute('id')
                                print(f"clicked slot with ID: {slot_id}")
                                click_nextbtn(driver)
                                kakute(driver)
                                pop_OK(driver)
                                cyucontinue(driver)
                               
                                if check_for_warning(driver):
                                    print("Warning detected, unchecking the slot")
                                    # Re-find the slot button before clicking to avoid stale element reference
                                    slot_button = driver.find_element(By.CSS_SELECTOR, f'input[id="{slot_id}"]')
                                    slot_button.click()  # Uncheck the slot
                                    continue  # Move to next slot
                                
                                
                            except (NoSuchElementException, ElementNotInteractableException) as e:
                                print(f"Failed to click slot within sat element, trying next...")
                                continue
                                return True  # Exit if all actions are successful
                    except NoSuchElementException as e:
                        print("Failed to find input elements within td element.")
                        continue
            except NoSuchElementException as e:
                print(f"Failed to find the time slot row for {time_slot_text}.")
                continue
        
    except (NoSuchElementException, ElementNotInteractableException) as e:
        print("Failed to find the 'sat' elements.")
        return False

def donini1521(driver):
    try:
        # 時間帯のテキストリスト
        time_slots = ["15:00 ～ 18:00", "18:00 ～ 21:00"]
        
        for time_slot_text in time_slots:
            try:
                # 時間帯を特定
                time_slot = driver.find_element(By.XPATH, f'//td[contains(text(), "{time_slot_text}")]/..')
                
                # <td> elements with class 'sat' within the specified time slot row
                sun_elements = time_slot.find_elements(By.CLASS_NAME, 'sun')
                
                for sun_element in sun_elements:
                    try:
                        # Find all <input> elements within the <td> element with an id that starts with 'i_record'
                        input_elements = sun_element.find_elements(By.CSS_SELECTOR, 'input[id^="i_record"]')
                        
                        for slot_button in input_elements:
                            try:
                                slot_button.click()
                                slot_id = slot_button.get_attribute('id')
                                print(f"clicked slot with ID: {slot_id}")
                                click_nextbtn(driver)
                                kakute(driver)
                                pop_OK(driver)
                                cyucontinue(driver)

                                if check_for_warning(driver):
                                    print("Warning detected, unchecking the slot")
                                    # Re-find the slot button before clicking to avoid stale element reference
                                    slot_button = driver.find_element(By.CSS_SELECTOR, f'input[id="{slot_id}"]')
                                    slot_button.click()  # Uncheck the slot
                                    continue  # Move to next slot
                                    
                            except (NoSuchElementException, ElementNotInteractableException) as e:
                                print(f"Failed to click slot within sun element, trying next...")
                                continue
                                return True  # Exit if all actions are successful
                    except NoSuchElementException as e:
                        print("Failed to find input elements within td element.")
                        continue
            except NoSuchElementException as e:
                print(f"Failed to find the time slot row for {time_slot_text}.")
                continue
        
    except (NoSuchElementException, ElementNotInteractableException) as e:
        print("Failed to find the 'sun' elements.")
        return False

def back(driver):
    try:
        backbtn = driver.find_element(By.CSS_SELECTOR, 'a[onclick="javascript:isSubmitDataSet(\'Back\');return false;"]')
        backbtn.click()
        time.sleep(2)
    except (NoSuchElementException, ElementNotInteractableException) as e:
        print("Failed to click the Back button.")
        return False
    return True

def hol1521(driver):
    try:
        # 時間帯のテキストリスト
        time_slots = ["15:00 ～ 18:00", "18:00 ～ 21:00"]
        
        for time_slot_text in time_slots:
            try:
                # 時間帯を特定
                time_slot = driver.find_element(By.XPATH, f'//td[contains(text(), "{time_slot_text}")]/..')
                
                # <td> elements with class 'sat' within the specified time slot row
                hol_elements = time_slot.find_elements(By.CLASS_NAME, 'hol')
                
                for hol_element in hol_elements:
                    try:
                        # Find all <input> elements within the <td> element with an id that starts with 'i_record'
                        input_elements = hol_element.find_elements(By.CSS_SELECTOR, 'input[id^="i_record"]')
                        
                        for slot_button in input_elements:
                            try:
                                slot_button.click()
                                slot_id = slot_button.get_attribute('id')
                                print(f"clicked slot with ID: {slot_id}")
                                click_nextbtn(driver)
                                kakute(driver)
                                pop_OK(driver)
                                cyucontinue(driver)

                                if check_for_warning(driver):
                                    print("Warning detected, unchecking the slot")
                                    # Re-find the slot button before clicking to avoid stale element reference
                                    slot_button = driver.find_element(By.CSS_SELECTOR, f'input[id="{slot_id}"]')
                                    slot_button.click()  # Uncheck the slot
                                    continue  # Move to next slot
                                    
                                
                            except (NoSuchElementException, ElementNotInteractableException) as e:
                                print(f"Failed to click slot within hol element, trying next...")
                                continue
                                return True  # Exit if all actions are successful
                    except NoSuchElementException as e:
                        print("Failed to find input elements within td element.")
                        continue
            except NoSuchElementException as e:
                print(f"Failed to find the time slot row for {time_slot_text}.")
                continue
        
    except (NoSuchElementException, ElementNotInteractableException) as e:
        print("Failed to find the 'hol' elements.")
        return False

def cyusyo(driver):
    try:
        # 待機して要素を見つける
        wait = WebDriverWait(driver, 2)
        cyusyobtn = wait.until(EC.element_to_be_clickable((By.XPATH, '//td[contains(text(), "抽選照会確定")]/..')))
        cyusyobtn.click()
        print("Clicked '抽選照会確定' button")
    except TimeoutException:
        print("Timed out waiting for the '抽選照会確定' button to be clickable")
    except NoSuchElementException:
        print("Failed to find the '抽選照会確定' button")


###各体育館のボダン###
#浪速スポーツセンター#
def naniwa2(driver):
    try:
        nariwa2 = driver.find_element(By.CSS_SELECTOR, 'tr[onclick="chng_chkboxImage(\'r_record17\', \'271004_001_22_02_01_002_001\', \'\', \'i_record17\', \'/osakashi/img/osakashi/\');"]')
        nariwa2.click()
        time.sleep(2)
    except (NoSuchElementException, ElementNotInteractableException) as e:
        print("Failed to click the naniwa button.")
        return False
    return True

#都島スポーツセンター#
def miya(driver):
    try:
        miyakojima2 = driver.find_element(By.CSS_SELECTOR, 'tr[onclick="chng_chkboxImage(\'r_record5\', \'271004_001_14_02_01_002_001\', \'\', \'i_record5\', \'/osakashi/img/osakashi/\');"]')
        miyakojima2.click()
        time.sleep(2)
    except (NoSuchElementException, ElementNotInteractableException) as e:
        print("Failed to click the miyakojima button.")
        return False
    return True

#淀川スポーツセンター＃
def yodo(driver):
    try:
        yodo2 = driver.find_element(By.CSS_SELECTOR, 'tr[onclick="chng_chkboxImage(\'r_record20\', \'271004_001_24_02_01_002_001\', \'\', \'i_record20\', \'/osakashi/img/osakashi/\');"]')
        yodo2.click()
        time.sleep(2)
    except (NoSuchElementException, ElementNotInteractableException) as e:
        print("Failed to click the yodo button.")
        return False
    return True
```
